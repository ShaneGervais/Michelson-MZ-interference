```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import c
```

```python
wavelength = 640e-9 #nm
largeur = 10e-9 #ns
MIN  = -5*largeur
MAX = 5*largeur
N = 1000
angle_pre_selection = np.pi/4
ellipse_pre = 0
phi_x = 0
phi_y = 0


d = 0 #diff√©rence de distance parcouru des deux faisceau (m)
side = 1e-5
spacing = side/N
k = 2*np.pi/wavelength
x1 = side/2 + d/2
y1 = side/2
x2 = side/2 - d/2
y2 = side/2


n = 1 #nombre de figure
```

```python
xi = np.empty([N,N])
freq = c/wavelength
w_0 = 2*np.pi*freq

for i in range(N):
    y=spacing*i
    for j in range(N):
        x=spacing*j
        r1 = np.sqrt(np.square(x-x1) + np.square(y-y1))
        r2 = np.sqrt(np.square(x-x2) + np.square(y-y2))
        xi[i,j] = np.cos(k*r1) + np.cos(k*r2)

plt.imshow(xi, origin="lower", extent=[0, side, 0, side], cmap='viridis')
plt.show()
```
